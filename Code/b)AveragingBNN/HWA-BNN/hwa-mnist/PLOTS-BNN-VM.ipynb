# Plots

```python
import numpy as np
import pandas as pd 
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import torch
```

```python
#Training
sgd = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_optsgd_lr0.01_bs128_avgp10.txt'
hwa10 = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp10.txt'
hwa50 = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp50.txt'
hwa100 = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp100.txt'


####Cyclical variant
sgd = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_optsgd_lr0.01_bs128_avgp10_cyclical.txt'
hwa10 = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp10_cyclical.txt'
hwa50 = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp50_cyclical.txt'
hwa100 = 'checkpoint/checkpoint_MNISTVM/trainlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp100_cyclical.txt'
```

```python
runsgd = pd.read_table(sgd, index_col = 0, sep = '\t')
runhwa10 = pd.read_table(hwa10, index_col = 0, sep = '\t')
runhwa50 = pd.read_table(hwa50, index_col = 0, sep = '\t')
runhwa100 = pd.read_table(hwa100, index_col = 0, sep = '\t')

num_epochs=50
every = len(runsgd['Train Loss'])/num_epochs

#xaxis = range(len(runhwa50['Train Loss']))
xaxis = range(num_epochs)
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Train Loss'].iloc[::every], label='SGD')
plt.plot(xaxis, runhwa10['Train Loss'].iloc[::every], label='HWA 10')
plt.plot(xaxis, runhwa50['Train Loss'].iloc[::every], label='HWA 50')
plt.plot(xaxis, runhwa100['Train Loss'].iloc[::every], label='HWA 100')
plt.ylim(22.5, 24)
plt.title('Train Loss')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Train Acc.'].iloc[::every], label='SGD')
plt.plot(xaxis, runhwa10['Train Acc.'].iloc[::every], label='HWA 10')
plt.plot(xaxis, runhwa50['Train Acc.'].iloc[::every], label='HWA 50')
plt.plot(xaxis, runhwa100['Train Acc.'].iloc[::every], label='HWA 100')
plt.ylim(0.8, 1)
plt.title('Train Acc.')
plt.grid(True)
plt.legend()
```

```python
#Testing!
sgd = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_optsgd_lr0.01_bs128_avgp10.txt'
hwa10 = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp10.txt'
hwa50 = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp50.txt'
hwa100 = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp100.txt'

###cyclical variant
sgd = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_optsgd_lr0.01_bs128_avgp10_cyclical.txt'
hwa10 = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp10_cyclical.txt'
hwa50 = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp50_cyclical.txt'
hwa100 = 'checkpoint/checkpoint_MNISTVM/testlogLeNet5_opthwa_sgd_lr0.01_bs128_avgp100_cyclical.txt'
```

```python
runsgd = pd.read_table(sgd, index_col = 0, sep = '\t')
runhwa10 = pd.read_table(hwa10, index_col = 0, sep = '\t')
runhwa50 = pd.read_table(hwa50, index_col = 0, sep = '\t')
runhwa100 = pd.read_table(hwa100, index_col = 0, sep = '\t')

num_epochs=50
every = len(runsgd['Test Loss'])/num_epochs

#xaxis = range(len(runhwa50['Train Loss']))
xaxis = range(num_epochs)
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Test Loss'].iloc[::every], label='SGD')
plt.plot(xaxis, runhwa10['Test Loss'].iloc[::every], label='HWA 10')
plt.plot(xaxis, runhwa50['Test Loss'].iloc[::every], label='HWA 50')
plt.plot(xaxis, runhwa100['Test Loss'].iloc[::every], label='HWA 100')

plt.title('Test Loss')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Test Acc.'].iloc[::every], label='SGD')
plt.plot(xaxis, runhwa10['Test Acc.'].iloc[::every], label='HWA 10')
plt.plot(xaxis, runhwa50['Test Acc.'].iloc[::every], label='HWA 50')
plt.plot(xaxis, runhwa100['Test Acc.'].iloc[::every], label='HWA 100')
plt.title('Test Acc.')
plt.grid(True)
plt.legend()
```
