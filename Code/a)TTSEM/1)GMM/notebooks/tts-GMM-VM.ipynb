```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import tensorflow as tf
import pandas as pd
import pickle
#ed.set_seed(314159)
```

```python
singlerun = pd.read_csv('filefromVM/runtts.csv')
vrsaem = pd.read_csv('vrsaem.csv')
```

```python
emrun = singlerun[singlerun.algo=='EM'].mu1.tolist()
iemrun = singlerun[singlerun.algo=='IEM'].mu1.tolist()
saemrun = singlerun[singlerun.algo=='saem'].mu1.tolist()
isaemrun = singlerun[singlerun.algo=='isaem'].mu1.tolist()
isaemsaga = singlerun[singlerun.algo=='isaemsaga'].mu1.tolist()
isaemvr = vrsaem[vrsaem.algo=='isaemvr'].mu1.tolist()
```

```python
xaxisrun = singlerun[singlerun.algo=='isaem'].iteration
xaxisrunem = singlerun[singlerun.algo=='EM'].iteration
xaxisrunvr = vrsaem[vrsaem.algo=='isaemvr'].iteration
```

```python
#5 RUNS
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 4.5))
axes.set_yscale('log')
plt.plot(xaxisrunem, emrun, label='Batch EM', marker='')
plt.plot(xaxisrunem, saemrun, label='Batch', marker='')
plt.plot(xaxisrun, iemrun, label='Inc. EM', marker='')
plt.plot(xaxisrun, isaemrun, label='Inc.', marker='')
plt.plot(xaxisrun, isaemsaga, label='FI', marker='')
plt.plot(xaxisrunvr, isaemvr, label='VR', marker='')

leg = plt.legend(fontsize=18,fancybox=True, loc=0,ncol=2)
leg.get_frame().set_alpha(0.5)
plt.xlabel('Epoch', fontsize=20)
plt.ylabel(r'$|\mu-\mu^*|^2$', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(linestyle='dotted',linewidth=2)
fig.tight_layout()
plt.show()
```

```python
#5 RUNS
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 4.5))
axes.set_yscale('log')
plt.plot(xaxisrunem*3, emrun, label='EM', marker='')
plt.plot(xaxisrunem*3, saemrun, label='SAEM', marker='')
plt.plot(xaxisrun*3, iemrun, label='iEM', marker='')
plt.plot(xaxisrun*3, isaemrun, label='iSAEM', marker='')
plt.plot(xaxisrun*3, isaemsaga, label='fiSAEM', marker='')
plt.plot(xaxisrunvr*3, isaemvr, label='vrSAEM', marker='')

leg = plt.legend(fontsize=18,fancybox=True, loc=0,ncol=2)
leg.get_frame().set_alpha(0.5)
plt.xlabel('Epoch', fontsize=20)
plt.ylabel(r'$|\mu-\mu^*|^2$', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(linestyle='dotted',linewidth=2)
fig.tight_layout()
plt.show()
```
