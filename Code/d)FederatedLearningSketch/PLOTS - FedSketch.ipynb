# Plots

```python
import numpy as np
import pandas as pd 
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import torch
```

```python
sgd = 'checkpoints/checkpoints_mnist/logcnn_optsgd_lr0.01_bs10_clients100_frac0.1.txt'
fedadam = 'checkpoints/checkpoints_mnist/logcnn_optfedadam_lr0.01_bs10_clients100_frac0.1.txt'

runsgd = pd.read_table(sgd, index_col = 0, sep = '\t')
runfedadam = pd.read_table(fedadam, index_col = 0, sep = '\t')
```

```python
xaxis = range(len(runsgd['Train Acc.']))
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Train Loss'], label='SGD')
plt.plot(xaxis, runfedadam['Train Loss'], label='FEDADAM')
plt.plot(xaxis, runsgd['Avg. Loss'], label='SGD', linestyle='dashed')
plt.plot(xaxis, runfedadam['Avg. Loss'], label='FEDADAM', linestyle='dashed')
plt.title('Training Loss')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Test Loss'], label='SGD')
plt.plot(xaxis, runfedadam['Test Loss'], label='FEDADAM')
plt.title('Testing Loss')
plt.grid(True)
plt.legend()
```

```python
xaxis = range(len(runsgd['Train Acc.']))
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Train Acc.'], label='SGD')
plt.plot(xaxis, runfedadam['Train Acc.'], label='FEDADAM')
plt.title('Training Accuracy')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Test Acc.'], label='SGD')
plt.plot(xaxis, runfedadam['Test Acc.'], label='FEDADAM')
plt.title('Testing Accuracy')
plt.grid(True)
plt.legend()
```
