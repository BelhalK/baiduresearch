```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import tensorflow as tf
#from tensorflow.examples.tutorials.mnist import input_data
#from edward.models import Categorical, Normal
#import edward as ed
import pandas as pd
import pickle
#ed.set_seed(314159)
```

```python
singlerun = pd.read_csv('pk500_mcstudy2.csv')
```

```python
saemrun = singlerun[singlerun.algo=='Batch'].V2.tolist()
isaemrun = singlerun[singlerun.algo=='Inc.'].V2.tolist()
isaemsaga = singlerun[singlerun.algo=='Fast Inc.'].V2.tolist()
isaemvr = singlerun[singlerun.algo=='VR'].V2.tolist()
```

```python
xaxisrun = singlerun[singlerun.algo=='Inc.'].iterations
xaxisrunem = singlerun[singlerun.algo=='Batch'].iterations
```

```python
#5 RUNS
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(8, 4.5))
plt.plot(xaxisrunem, saemrun, label='SAEM', marker='')
plt.plot(xaxisrun, isaemrun, label='iSAEM', marker='')
plt.plot(xaxisrun, isaemsaga, label='fiSAEM', marker='')
plt.plot(xaxisrun, isaemvr, label='vrSAEM', marker='')

leg = plt.legend(fontsize=18,fancybox=True, loc=0,ncol=2)
leg.get_frame().set_alpha(0.5)
plt.xlabel('Epoch', fontsize=20)
plt.ylabel(r'$|k_a-k_a^*|^2$', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(linestyle='dotted',linewidth=2)
fig.tight_layout()
plt.show()
```

```python
fig.savefig("pk500.png",bbox_inches = 'tight')
```
