# Plots

```python
import numpy as np
import pandas as pd 
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import torch
```

```python
filediag = 'checkpoint/checkpoint_MNIST/logMnistNetSmall_diagnostic.txt'
run = pd.read_table(filediag, index_col = 0, sep = '\t')
xaxis = range(len(run['Train Loss']))
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, run['Train Loss'], label='Train loss')
plt.plot(xaxis, run['Train Acc.'], label='Train accuracy')
plt.plot(xaxis, run['iploss'], label='Inner Prod', linestyle='dashed')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, run['newstat'], label='New Statistic', linestyle='dashed')
plt.plot(xaxis, list(run.index), label='LR', linestyle='dashed')
plt.grid(True)
plt.legend()
```

```python
with open ('params/MnistNet/trained_weights', 'rb') as fp:
    trained_weights = torch.load(fp)
with open ('params/MnistNet/init_weights', 'rb') as fp:
    init_weights = torch.load(fp)
init_weights = init_weights.detach().numpy()
```

```python
weights = list(map(list, zip(*trained_weights)))
```

```python
scalarindex = 2 #between 0 and 9 (since fc1 output)
weights = []
for element in trained_weights:
        weights.append(element[scalarindex])
weights = list(map(list, zip(*weights)))
```

```python
len(weights)
```

```python
dim = 3 #scalar weight index
for dim in range(200,203):
    distanceweights = [x - init_weights[scalarindex][dim] for x in weights[dim]] 
    xaxis = range(1,len(distanceweights)+1)
    plt.plot(xaxis, distanceweights)
    plt.xlabel('iterations')
    plt.ylabel(r'$|w_d - w_d^0|$', fontsize='small')
    plt.grid(True)
    plt.legend()
```

```python
dim = 3 #scalar weight index
for dim in range(500,730):
    distanceweights = [x - init_weights[scalarindex][dim] for x in weights[dim]] 
    xaxis = [element/4 for element in range(1,len(distanceweights)+1)]
    plt.plot(xaxis, distanceweights)
    plt.xlabel('iterations')
    plt.ylabel(r'$|w_d - w_d^0|$', fontsize='small')
    plt.grid(True)
    plt.legend()
```

```python
dim = 3 #scalar weight index
for dim in range(700,730):
    distanceweights = [x - init_weights[scalarindex][dim] for x in weights[dim]] 
    xaxis = [element/3.3 for element in range(1,len(distanceweights)+1)]
    plt.plot(xaxis, distanceweights)
    plt.xlabel('iterations')
    plt.ylabel(r'$|w_d - w_d^0|$', fontsize='small')
    plt.grid(True)
    plt.legend()
```

```python
def mean(a):
    return sum(a) / len(a)

scalarindex = 9
distanceweights=[]
N_start = 700
N_end = 710
for dim in range(N_start, N_end):
    distanceweights.append([(x - init_weights[scalarindex][dim])**2 for x in weights[dim]] )
distanceweights = map(mean, zip(*distanceweights))
xaxis = [element/3.3 for element in range(1,len(distanceweights)+1)]

plt.plot(xaxis, distanceweights)
plt.xlabel('iterations')
plt.ylabel(r'$|w_d - w_d^0|$', fontsize='small')
plt.grid(True)
plt.legend()
```
