# Plots

```python
import numpy as np
import pandas as pd 
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
import torch
```

### Training

```python
sgd = 'checkpoint/checkpoint_MNIST/trainlogLeNet5_optsgd_lr0.01_bs128.txt'
hwa10 = 'checkpoint/checkpoint_MNIST/trainlogLeNet5_opthwa_lr0.001_bs128_avgp10.txt'
hwa50 = 'checkpoint/checkpoint_MNIST/trainlogLeNet5_opthwa_lr0.001_bs128_avgp50.txt'

runsgd = pd.read_table(sgd, index_col = 0, sep = '\t')


xaxis = range(len(runsgd['Train Loss']))
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Train Loss'], label='SGD')

plt.title('Train Loss')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Train Acc.'], label='SGD')
plt.title('Train Acc.')
plt.grid(True)
plt.legend()
```

### Testing

```python
sgd = 'checkpoint/checkpoint_MNIST/testlogLeNet5_optsgd_lr0.01_bs128.txt'
hwa10 = 'checkpoint/checkpoint_MNIST/testlogLeNet5_opthwa_lr0.001_bs128_avgp10.txt'
hwa50 = 'checkpoint/checkpoint_MNIST/testlogLeNet5_opthwa_lr0.001_bs128_avgp50.txt'

runsgd = pd.read_table(sgd, index_col = 0, sep = '\t')


xaxis = range(len(runsgd['Test Loss']))
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Test Loss'], label='SGD')

plt.title('Train Loss')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Test Acc.'], label='SGD')
plt.title('Train Acc.')
plt.grid(True)
plt.legend()
```

```python
sgd = 'checkpoint/checkpoint_CIFARVM/trainlogResNet_optsgd_lr0.001_bs128_avgp10.txt'
hwa10 = 'checkpoint/checkpoint_CIFARVM/trainlogResNet_opthwa_sgd_lr0.001_bs128_avgp10.txt'
hwa50 = 'checkpoint/checkpoint_CIFARVM/trainlogResNet_opthwa_sgd_lr0.001_bs128_avgp50.txt'
hwa100 = 'checkpoint/checkpoint_CIFARVM/trainlogResNet_opthwa_sgd_lr0.001_bs128_avgp100.txt'

runsgd = pd.read_table(sgd, index_col = 0, sep = '\t')
runhwa10 = pd.read_table(hwa10, index_col = 0, sep = '\t')
runhwa50 = pd.read_table(hwa50, index_col = 0, sep = '\t')
runhwa100 = pd.read_table(hwa100, index_col = 0, sep = '\t')


xaxis = range(len(runsgd['Train Loss']))
f, axs = plt.subplots(2,2,figsize=(10,5))
plt.tight_layout()
plt.subplot(1, 2, 1)
plt.plot(xaxis, runsgd['Train Loss'], label='SGD')
plt.plot(xaxis, runhwa10['Train Loss'], label='HWA10')
plt.plot(xaxis, runhwa50['Train Loss'], label='HWA50')

plt.title('Train Loss')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(xaxis, runsgd['Train Acc.'], label='SGD')
plt.plot(xaxis, runhwa10['Train Acc.'], label='HWA10')
plt.plot(xaxis, runhwa50['Train Acc.'], label='HWA50')
plt.title('Train Acc.')
plt.grid(True)
plt.legend()
```
